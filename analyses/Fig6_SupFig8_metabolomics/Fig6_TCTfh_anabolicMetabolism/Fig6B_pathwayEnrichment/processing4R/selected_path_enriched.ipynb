```python
new_df = pd.DataFrame()
for x in folder_l:
    temp = pd.read_csv("../X0_raw_pval_style/enrichmen/" + x + "/" + "msea_ora_result.csv")
    temp["group"] = x #.split("__")[1]
    temp = temp.loc[0:9,:]
    new_df = pd.concat([new_df, temp])
```

```python
#FDR < 0.25
sel_new_df = new_df.loc[(new_df["FDR"] < 0.25)  | 
                        ((new_df["Raw p"] < 0.05) & (new_df['group']=="TC_TnvsB6_Tn")),:]
```
